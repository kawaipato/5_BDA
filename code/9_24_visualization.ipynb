### 베트남 관광객 추이 그래프 시각화

```python
visit = pd.read_csv('/Users/seoihwan/Desktop/escuela/BDA_proj/5_BDA/data/visitors.csv')
viet = visit[visit['국적별'] == '베트남']
viet
```

```python
vietnam = viet.drop(columns=viet.columns[1:10],axis=1).T
vietnam.columns = ['베트남']
vietnam = vietnam.iloc[1:-2,]
vietnam['베트남'] = vietnam['베트남'].str.replace(',','').astype('int')
vietnam['rank'] = [14,11,10,8,8,7,7,9,5,5]
```

```python
from PIL import Image
fig = plt.figure(figsize=(10,8)) ## Figure 생성 
fig.set_facecolor('white') ## Figure 배경색 지정
ax1 = fig.add_subplot() ## axes 생성

# colors = sns.color_palette('Reds', len(vietnam)) ## 바 차트 색상

xtick_label_position = list(range(len(vietnam))) ## x축 눈금 라벨이 표시될 x좌표
ax1.set_xticks(xtick_label_position) ## x축 눈금 
ax1.set_xticklabels(vietnam.index) ## x축 눈금 라벨
ax1.bar(xtick_label_position, vietnam['베트남'], color="#FF4D4D") ## 바차트 출력
# ax1.set_ylabel('방문객 수')

color = "#FFC300"
ax2 = ax1.twinx() ## 새로운 axis 생성
ax2.plot(xtick_label_position, vietnam['rank'], color=color, linestyle='--', marker='o') ## 선 그래프 
ax2.tick_params(axis='y') ## 눈금 라벨 색상 지정
ax2.set_ylim(1,15)
# ax2.set_ylabel('순위',rotation=360)
plt.gca().invert_yaxis()
# ax1.set_xlabel('년도',)

# plt.grid(linestyle='--', linewidth=0.5, alpha=0.5)
# plt.title('베트남 방한 관광객 추이', fontsize=20)
plt.show()
```

### 베트남 관광객 증감율 시각화

```python
vietnam['전년도_방문자수'] = vietnam['베트남'].shift(1)
vietnam['전년대비증가율'] = ((vietnam['베트남'] - vietnam['전년도_방문자수'])/vietnam['전년도_방문자수'])*100
vietnam
```

```python
vietnam.to_csv('베트남 관광객 추이.csv',index=False)
```
